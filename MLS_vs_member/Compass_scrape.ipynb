```python

def extract_house(page):
    house=f'https://www.compass.com/homes-for-sale/grayson-county-tx/locations=41016,40712,41827,41267,41253,42366,41528,42137,42054,41750,41640,42618,42592,42314,41088,42407,41351,41365,40726,41203,42164/property-type=single-family/start={page}/'

    url = house
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36',
            "Accept-Encoding": "*",
    "Connection": "keep-alive"}
    r = requests.get(url, headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    return soup

def transform_house(soup,type1):
    div1 = soup.find_all('div', class_= 'uc-listingCard-leftWrapper')
    for item in div1:
        addr = item.find('a').text
        city = item.find('h2').text
        price = item.find('div').text        
        mls = {
            'addr':addr,
            'full_addr':city,
            'price':price,
            'type':type1}
        mlslist_house.append(mls)
    return
mlslist_house = []
```

```python


def transform_condo(soup,type1):
    div1 = soup.find_all('div', class_= 'uc-listingCard-leftWrapper')
    for item in div1:
        addr = item.find('a').text
        city = item.find('h2').text
        price = item.find('div').text        
        mls = {
            'addr':addr,
            'full_addr':city,
            'price':price,
            'type':type1}
        mlslist_condo.append(mls)
    return
mlslist_condo = []

def transform_multi(soup,type1):
    div1 = soup.find_all('div', class_= 'uc-listingCard-leftWrapper')
    for item in div1:
        addr = item.find('a').text
        city = item.find('h2').text
        price = item.find('div').text        
        mls = {
            'addr':addr,
            'full_addr':city,
            'price':price,
            'type':type1}
        mlslist_multi.append(mls)
    return
mlslist_multi = []

def transform_townhouse(soup,type1):
    div1 = soup.find_all('div', class_= 'uc-listingCard-leftWrapper')
    for item in div1:
        addr = item.find('a').text
        city = item.find('h2').text
        price = item.find('div').text        
        mls = {
            'addr':addr,
            'full_addr':city,
            'price':price,
            'type':type1}
        mlslist_townhouse.append(mls)
    return
mlslist_townhouse = []
```

```python
def transform2(soup):
    div1 = soup.find_all('span', class_= 'Container-sc-133nrr9')
    for item in div1:
        cnt = item.find('strong', {'data-tn':'resultsBar-totalItemsCount'}).text
    return cnt
# mlslist = []
```

```python
#HTML code extract
c1 = extract_house(21)
c2 = extract_condo(21)
c3 = extract_multi(21)
c4 = extract_townhouse(21)

#count pages of listing types
cnt_house = transform2(c1).replace(',','')
cnt_condo = transform2(c2).replace(',','')
cnt_multi = transform2(c3).replace(',','')
cnt_townhouse = transform2(c4).replace(',','')

#setting recurring page by page counts
pages_house = int(int(cnt_house)/21)
pages_condo = int(int(cnt_condo)/21)
pages_multi = int(int(cnt_multi)/21)
pages_townhouse = int(int(cnt_townhouse)/21)

print(pages_house,pages_condo,pages_multi,pages_townhouse)
```
